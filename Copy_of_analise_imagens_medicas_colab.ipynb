## 4.5 Indicadores de desempenho

O Keras não possui os indicadores de desempenho como precisão, sensibilidade e pontuação f1 por padrão, portanto precisamos implementar externamente.

````python
def recall_score(y_true, y_pred):
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    possible_positives = K.sum(K.round(K.clip(y_true, 0, 1)))
    recall = true_positives / (possible_positives + K.epsilon())
    return recall

def precision_score(y_true, y_pred):
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    predicted_positives = K.sum(K.round(K.clip(y_pred, 0, 1)))
    precision = true_positives / (predicted_positives + K.epsilon())
    return precision

def f1_score(y_true, y_pred):
    precision = precision_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    return 2*((precision*recall)/(precision+recall+K.epsilon()))
````

### 4.5.1 Arquitetura ResNet50

**Pergunta**: Explique como é constituída a arquitetura do ResNet50? *Utilize, se necessário, gráficos, projetos que utilizam essa arquitetura. Detalhe também sua topologia em camadas e mostre quais as situações essa arquitetura pode ter mais êxito e quais cenários não tem.*

**Resposta**:  A ResNet-50 é uma rede neural convolucional com 50 camadas de profundidade. Onde se pode carregar uma versão pré-treinada. A rede pré-treinada pode classificar imagens em 1000 categorias de objetos, como teclado, mouse, lápis e muitos animais. Como resultado, a rede aprende representações de recursos ricas para uma ampla gama de imagens. A rede tem um tamanho de entrada de imagem de 224 por 224. 


arquitetura-restnet-50.png


Acima uma imagem de como funciona a arquitetura da RestNet50

A ResNet foi construída por blocos residuais, estes blocos tem uma entrada x que passa por uma série de operações de convolução-relu-convolução. O resultado da operação f(x) é adicionado à entrada original x.
Essas camadas adicionadas são o mapeamento de identidade e as outras camadas são copiadas do modelo mais raso aprendido. Esta solução indica que um modelo mais profundo não deve produzir um erro de treinamento maior do que sua versão da rede mais rasa .
A equação 2 mostra o processamento que o bloco residual faz.
H(x)= f(x)+x


bloco-residual.png

Bloco residual da rede RestNet50



A técnica de transfer learning consiste de utilizar o mesmo modelo e treiná-lo para outas imagens. Por tal motivo, excluímos a última camada para modelar com as classes que definimos, ou seja, **controle**, **bacteriana** e **viral**.

Informe a quantidade de classes a serem classificadas.

````python
## IMPLEMENTE

qtde_classes = 3
````

````python
conv_base = ResNet50(include_top=False)

for layer in conv_base.layers:
    layer.trainable = False
````

````python
x = conv_base.output
x = layers.GlobalAveragePooling2D()(x)
x = layers.Dense(128, activation='relu')(x) 

predictions = layers.Dense(qtde_classes, activation='softmax')(x)
model = Model(conv_base.input, predictions)

model.summary()
````

````python
optimizer = optimizers.Adam()
model.compile(loss='categorical_crossentropy', optimizer=optimizer, metrics=[precision_score, recall_score, f1_score])
````

O número de épocas define quantas vezes o modelo irá treinar e validar o erro, assim ajustando os pesos para melhor convergência.
Escolha o número adequado de épocas para alcançarmos pelo menos **70% de precisão de validação**.

````python
## IMPLEMENTE

qtde_epocas = 10
````

````python
history = model.fit(train_generator, epochs=qtde_epocas, validation_steps=5, steps_per_epoch=5, 
                    validation_data=val_generator)
````

Um modelo que converge bem possui o gráfico de perda (*loss*) descendente e os gráfico de precisão (*precision*), sensibilidade (*recall*) e pontuação f1 (*f1 score*) em acendente.

````python
# Exibindo dados de Precisão
plt.plot(history.history['precision_score'])
plt.plot(history.history['val_precision_score'])
plt.title('model precision')
plt.ylabel('precision')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

# Exibindo dados de Sensibilidade
plt.plot(history.history['recall_score'])
plt.plot(history.history['val_recall_score'])
plt.title('model recall')
plt.ylabel('recall')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

# Exibindo dados de F1 Score
plt.plot(history.history['f1_score'])
plt.plot(history.history['val_f1_score'])
plt.title('model f1_score')
plt.ylabel('f1_score')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

# Exibindo dados de Perda
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
````

### 4.5.3 Arquitetura VGG19

**Resposta**:  Esta arquitetura, que foi uma das primeiras a aparecer, foi introduzida por Simonyan e Zisserman em 2014 com seu artigo intitulado Redes convolucionais muito profundas para reconhecimento de imagens em grande escala.
vgg19.png


É uma arquitetura simples, usando apenas blocos compostos por um número incremental de camadas convolucionais com filtros de tamanho 3x3. Além disso, para reduzir o tamanho dos mapas de ativação obtidos, blocos de máximo pooling são intercalados entre os convolucionais, reduzindo o tamanho desses mapas de ativação pela metade. Por fim, é utilizado um bloco de classificação, consistindo em duas camadas densas de 4.096 neurônios cada, e a última camada, que é a camada de saída, de 1.000 neurônios.

O 16 e o ​​19 se referem ao número de camadas ponderadas que cada rede possui (camadas convolucionais e densas, camadas de pooling não são contadas). Eles correspondem às colunas D e E na tabela abaixo.

vgg-19.png

Esta arquitetura foi proposta por François Chollet (o criador de Keras) e a única coisa que ele traz para a Origem é que ele faz as convoluções da maneira ideal para que demorem menos. Isso é alcançado separando as convoluções 2D em 2 convoluções 1D.  A arquitetura Xception e a combinacao da ideia do modulo da Inception  e do conceito de depthwise separable convolutions. Dada uma imagem de entrada de Iw×h×d, onde w representa o comprimento da imagem, h representa a altura e d representa o numero de canais, as depthwise separable convolutions realizam a etapa de convolucao em dois passos (ao inves de um unico como nas demais arquiteturas). No primeiro passo, e realizada a convolucao de um kernel de n × n × 1, onde n representa o tamanho do kernel, com todos os canais de I individualmente. Logo apos, no segundo passo da convolucao, os resultados do passo 1 sao combinados utilizando um novo kernel de 1 × 1 × d. Isso melhora o desempenho do processamento das convolucoes, possibilitando que a rede faca menos operacoes ao longo do treinamento.
